```python
import seaborn as sns  
import matplotlib.pyplot as plt  
import numpy as np
```

```python
# Ma'lumotlar (x va y o'qlari uchun sonlar)
x = [1, 2, 3, 4, 5]  
y = [2, 4, 6, 8, 10]  

# Grafik chizish
plt.plot(x, y)  

# Grafikni ko'rsatish
plt.show()
```

```python

# Ma'lumotlar
x = [1, 2, 3, 4, 5]  
y = [2, 4, 6, 8, 10]  

# Seaborn yordamida chizish
sns.lineplot(x=x, y=y)  

# Grafikni ko'rsatish
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

plt.plot(x, y, color='red', linestyle='--', marker='o', markersize=8, linewidth=2, label="Chiziq 1")
plt.xlabel("X o'qi")
plt.ylabel("Y o'qi")
plt.title("Oddiy Chiziqli Grafik")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = np.random.randn(1000)  # 1000 ta tasodifiy son

plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel("Qiymatlar")
plt.ylabel("Frekvensiya")
plt.title("Gistogramma")
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='purple', alpha=0.5, marker='*')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Nuqtaviy Grafik (Scatter Plot)")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 30, 25]

plt.bar(categories, values, color=['red', 'blue', 'green', 'purple', 'orange'])
plt.xlabel("Kategoriya")
plt.ylabel("Qiymat")
plt.title("Ustunli Diagramma")
plt.show()
```

```python
labels = ['Apple', 'Samsung', 'Huawei', 'Xiaomi']
sizes = [30, 40, 15, 15]
colors = ['red', 'blue', 'green', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Bozor Ulushlari")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z, c='r', marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
sns.set_theme(style="darkgrid")

x = np.linspace(0, 10, 100)
y = np.sin(x)

sns.lineplot(x=x, y=y, color="blue")
plt.title("Seaborn Chiziqli Grafik")
plt.show()
```

```python
data = np.random.randn(1000)

sns.histplot(data, bins=30, kde=True, color="red")
plt.title("Seaborn Gistogramma")
plt.show()
```

```python
data = np.random.randn(100)

sns.boxplot(data=data, color="green")
plt.title("Seaborn Box Plot")
plt.show()
```
