```python
import seaborn as sns  
import matplotlib.pyplot as plt  
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

plt.plot(x, y, color='red', linestyle='--', marker='o', markersize=8, linewidth=2, label="Chiziq 1")
plt.xlabel("X o'qi")
plt.ylabel("Y o'qi")
plt.title("Oddiy Chiziqli Grafik")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = np.random.randn(1000)  # 1000 ta tasodifiy son

plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel("Qiymatlar")
plt.ylabel("Frekvensiya")
plt.title("Gistogramma")
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='purple', alpha=0.5, marker='*')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Nuqtaviy Grafik (Scatter Plot)")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 30, 25]

plt.bar(categories, values, color=['red', 'blue', 'green', 'purple', 'orange'])
plt.xlabel("Kategoriya")
plt.ylabel("Qiymat")
plt.title("Ustunli Diagramma")
plt.show()
```

```python
labels = ['Apple', 'Samsung', 'Huawei', 'Xiaomi']
sizes = [30, 40, 15, 15]
colors = ['red', 'blue', 'green', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Bozor Ulushlari")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 4))  # Grafik o‘lchami

plt.subplot(1, 2, 1)  # 1-qator, 2-ustun, 1-chi grafik
plt.plot(x, y1, 'r')
plt.title("Sinus")

plt.subplot(1, 2, 2)  # 1-qator, 2-ustun, 2-chi grafik
plt.plot(x, y2, 'b')
plt.title("Kosinus")

plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))  # 2x2 grafiklar

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax[0, 0].plot(x, y, 'r')
ax[0, 0].set_title("Grafik 1")

ax[0, 1].plot(x, -y, 'g')
ax[0, 1].set_title("Grafik 2")

ax[1, 0].plot(x, y**2, 'b')
ax[1, 0].set_title("Grafik 3")

ax[1, 1].plot(x, np.sqrt(x), 'm')
ax[1, 1].set_title("Grafik 4")

plt.tight_layout()  # Grafiklar orasidagi masofani to‘g‘rilash
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, 'b')
plt.text(5, 0, "Bu Sinus Funktsiyasi", fontsize=12, color='red')
plt.annotate("Maksimum Nuqta", xy=(np.pi/2, 1), xytext=(2, 0.8),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
x = np.linspace(1, 100, 100)
y = np.exp(x/20)

plt.plot(x, y, 'r')
plt.yscale("log")  # Logarifmik o‘lchov
plt.xlabel("X")
plt.ylabel("Y (Logarifmik)")
plt.title("Logarifmik Grafik")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.title("3D Yuzali Grafik")
plt.show()
```

```python
data = np.random.rand(10, 10)  # 10x10 matritsa
sns.heatmap(data, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Issiqlik Xarita")
plt.show()
```

```python
data = np.random.randn(1000)
sns.kdeplot(data, shade=True, color="green")
plt.title("Zichlik Grafik")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex", palette="Set1")
plt.title("Strip Plot")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
plt.title("Violin Plot")
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Regressiya Chizig'i")
plt.show()
```

```python
sns.pairplot(tips, hue="sex")
plt.show()
```
